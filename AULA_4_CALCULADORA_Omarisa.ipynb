```python
historico = []
continuar = "s"

while continuar == "s":
    print("\n--- Nova Conta ---")
    numero1 = int(input("Digite o primeiro número: "))
    numero2 = int(input("Digite o segundo número: "))
    operacao = input("Digite + - * ou / : ")

    if operacao == "+":
        resultado = numero1 + numero2
        conta = f"{numero1} + {numero2} = {resultado}"
        historico.append(conta)
        print("Resultado:", resultado)

    elif operacao == "-":
        resultado = numero1 - numero2
        conta = f"{numero1} - {numero2} = {resultado}"
        historico.append(conta)
        print("Resultado:", resultado)

    elif operacao == "*":
        resultado = numero1 * numero2
        conta = f"{numero1} * {numero2} = {resultado}"
        historico.append(conta)
        print("Resultado:", resultado)

    elif operacao == "/":
        if numero2 == 0:
            print("Erro: Não pode dividir por zero!")
            conta = f"{numero1} / {numero2} = Erro"
            historico.append(conta)
        else:
            resultado = numero1 / numero2
            conta = f"{numero1} / {numero2} = {resultado}"
            historico.append(conta)
            print("Resultado:", resultado)
    else:
        print("Operação inválida!")

    continuar = input("\nQuer fazer outra conta? Digite s para sim: ")

print("\n--- Histórico de Contas ---")
for item in historico:
    print(item)

print("Calculadora encerrada. Você é braba!")
```
